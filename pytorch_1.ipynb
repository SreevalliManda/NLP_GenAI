```python
import pandas as pd

# URL of the CSV file
url = "https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(url)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df.head()
```

```python
df['diagnosis'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

```python
import torch
```

```python
X_train_tensor = torch.from_numpy(X_train).float()  # Convert to float32
X_test_tensor = torch.from_numpy(X_test).float()    # Convert to float32
y_train_tensor = torch.from_numpy(y_train).long()   # Convert to int64 (long)
y_test_tensor = torch.from_numpy(y_test).long()     # Convert to int64 (long)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1=nn.Linear(30,10)
        self.relu=nn.ReLU()
        self.fc2=nn.Linear(10,2)
        
    def forward(self,x):
        out=self.relu(self.fc1(x))
        out=self.fc2(out)

        return out
    
```

```python
model=Model()
```

```python

criterion=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python
for epoch in range(100):
    optimizer.zero_grad()
    outputs=model(X_train_tensor)
    loss=criterion(outputs,y_train_tensor)
    loss.backward()
    optimizer.step()
    print(f'epoch{epoch+1} : Loss:{loss.item()} ')
```

```python
model.eval()  # Set the model to evaluation mode
with torch.no_grad():
    test_outputs = model(X_test_tensor)
    _, predicted = torch.max(test_outputs, 1)
    print(f'predicted:{predicted}' )
    print(f'Actual:{y_test_tensor}')
    accuracy = (predicted == y_test_tensor).float().mean()
    print(f'Accuracy: {accuracy.item()}')
```
